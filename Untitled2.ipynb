```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import calinski_harabasz_score, silhouette_score
```

```python
X, _ = make_blobs(n_samples=300, n_features=5, centers=4, random_state=42)
```

```python
print(f"Taille des données: {X.shape}")
```

```python
def run_kmeans(X, init_method='random', n_clusters=4):
    """
    Exécute K-means avec différentes initialisations.
    
    Parameters:
    X (array): Données d'entrée
    init_method (str): Méthode d'initialisation
    n_clusters (int): Nombre de clusters
    
    Returns:
    y_pred (array): Étiquettes des clusters
    centers (array): Coordonnées des centres
    """
    kmeans = KMeans(n_clusters=n_clusters, init=init_method, random_state=42)
    y_pred = kmeans.fit_predict(X)
    centers = kmeans.cluster_centers_
    return y_pred, centers
```

```python
y_pred_random, centers_random = run_kmeans(X, init_method='random')
```

```python
y_pred_kmeanspp, centers_kmeanspp = run_kmeans(X, init_method='k-means++')
```

```python
def evaluate_clustering(X, y_pred):
    """
    Évalue le clustering avec Calinski-Harabasz et Silhouette.
    
    Parameters:
    X (array): Données d'entrée
    y_pred (array): Étiquettes des clusters
    
    Returns:
    ch_score (float): Score de Calinski-Harabasz
    silhouette_avg (float): Score de Silhouette
    """
    ch_score = calinski_harabasz_score(X, y_pred)
    silhouette_avg = silhouette_score(X, y_pred)
    return ch_score, silhouette_avg
```

```python
ch_score_random, silhouette_random = evaluate_clustering(X, y_pred_random)
ch_score_kmeanspp, silhouette_kmeanspp = evaluate_clustering(X, y_pred_kmeanspp)
```

```python
print(f"Random Init - Calinski-Harabasz Score: {ch_score_random}, Silhouette Score: {silhouette_random}")
print(f"K-means++ Init - Calinski-Harabasz Score: {ch_score_kmeanspp}, Silhouette Score: {silhouette_kmeanspp}")
```

```python
if ch_score_kmeanspp > ch_score_random and silhouette_kmeanspp > silhouette_random:
    print("K-means++ donne de meilleurs résultats.")
else:
    print("L'initialisation aléatoire donne de meilleurs résultats.")
```

```python
def plot_clusters(X, y_pred, centers):
    """
    Trace les clusters après réduction par PCA.
    
    Parameters:
    X (array): Données d'entrée
    y_pred (array): Étiquettes des clusters
    centers (array): Coordonnées des centres
    """
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)
    centers_pca = pca.transform(centers)
    plt.figure(figsize=(10, 6))
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_pred, cmap='viridis', alpha=0.5, label='Data points')
    plt.scatter(centers_pca[:, 0], centers_pca[:, 1], marker='o', s=200, c='red', edgecolors='k', label='Cluster centers')
    plt.title('Clustering with K-means')
    plt.legend()
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()
```

```python
plot_clusters(X, y_pred_kmeanspp, centers_kmeanspp)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print("Nouvelle matrice des observations:")
print(X_pca)
```

```python
print("Valeurs propres:", pca.explained_variance_)
print("Vecteurs propres (axes principaux):")
print(pca.components_)
```

```python
print("Inertie de chaque axe:", pca.explained_variance_ratio_)
```

```python
print("Somme des inerties:", np.sum(pca.explained_variance_ratio_))
```
